```python
df['Open'].plot()
df.rolling(window=30).mean()['Close'].plot(figsize=(16,6))
```

```python
df['Close 30 day MA'] = df['Close'].rolling(30).mean()
df[['Close 30 day MA', 'Close']].plot(figsize=(16,6))
```

```python
df['Close'].expanding(min_periods=1).mean().plot(figsize=(16,6))
# the average of all the values that came before it.
```

```python
# Bollinger bands are bands placed above and below the moving averages
# the bands expand as the bolinger bands expand
# used to determine whether a price movment is significant
```

```python
# create close 20 MA
df['Close: 20 Day Mean'] = df['Close'].rolling(20).mean()
# upper = 20MA + 2*std(20)
df['Upper'] = df['Close: 20 Day Mean'] + 2*(df['Close'].rolling(20).std())
# lower = 20MA - 2* std(20)
df['Lower'] = df['Close: 20 Day Mean'] - 2*(df['Close'].rolling(20).std())

# close
df[['Close', 'Close: 20 Day Mean', 'Upper', 'Lower']].plot(figsize=(16,6))
```
